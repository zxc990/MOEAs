```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9,9
                         ))
ax = plt.subplot(111, projection='3d')  # 
r = 0.5                     #
v = 0.0099                       #
l = math.floor(r / v)
parr = [0,1,2,3,4]
#parr = [0]
final_data = [[],[],[]]
for j in range(l):
    data = [[],[],[]]
    #5 5
    for i in range(len(parr)):
        tem_x = math.cos(math.pi *( 0.4 * parr[i]+0.45)) * r + 0.5
        tem_y = math.sin(math.pi *( 0.4 * parr[i]+0.45)) * r + 0.5
        data[0].append(math.cos(math.pi *( 0.4 * parr[i]+0.45)) * r + 0.5)
        data[1].append(math.sin(math.pi *( 0.4 * parr[i]+0.45)) * r + 0.5)
        data[2].append((- tem_x - tem_y + 1.7)/1.5)
        R = r * 0.5
        #math.sin(math.pi * 0.1)/math.sin(math.pi * 0.2)+ 0.5
        tem_x = math.cos(math.pi * 0.4 * parr[i] + math.pi * 0.65) * R + 0.5
        tem_y = math.sin(math.pi * 0.4 * parr[i] + math.pi * 0.65) * R + 0.5
        data[0].append(math.cos(math.pi * 0.4 * parr[i] + math.pi * 0.65) * R + 0.5)
        data[1].append(math.sin(math.pi * 0.4 * parr[i] + math.pi * 0.65) * R + 0.5)
        data[2].append((- tem_x - tem_y + 1.7)/1.5)
    r = r - v
    if r <= 0:
        break
    
    tdt = [[],[],[]]
    # 
    for i in range(len(data[0])):
        cur = i
        if i == len(data[0]) - 1:
            nex = 0
        else:
            nex = i + 1  

        k = (data[1][nex] - data[1][cur]) / (data[0][nex] - data[0][cur])
        b = data[1][nex] - k * data[0][nex]
        xs = np.arange(data[0][cur],data[0][nex],(data[0][nex] - data[0][cur]) / 20)        #20
        xs = xs[1:len(xs)]
        xy = []
        for j in range(len(xs)):
            tdt[0].append(xs[j])
            tdt[1].append(k * xs[j] + b)
            tdt[2].append((- xs[j] - k * xs[j] - b + 1.7)/1.5)
#     x, y, z = data[0],data[1],data[2]       #r10
#     ax.scatter(x[:len(x)], y[:len(y)], z[:len(z)], c='b',s=1)  # 
#     x, y, z = tdt[0],tdt[1],tdt[2]          #r
#     ax.scatter(x[:len(x)], y[:len(y)], z[:len(z)], c='b',s=1)  # 
    temp_final0 = data[0] + tdt[0]
    temp_final1 = data[1] + tdt[1]
    temp_final2 = data[2] + tdt[2]
    for i in range(len(temp_final0)):
        final_data[0].append(temp_final0[i])
        final_data[1].append(temp_final1[i])
        final_data[2].append(temp_final2[i])
x, y, z = final_data[0],final_data[1],final_data[2]          #r
ax.scatter(x, y, z, c='b',s=1)  #   
ax.view_init(elev=45, azim=45)
```
